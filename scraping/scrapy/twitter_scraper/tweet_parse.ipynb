```python
mentions = []
for i in range(len(data)):
    mention = data.iloc[i].entities['user_mentions']
    if mention == []:
        mentions.append('NA')
    else:
        submen = []
        for j in range(len(mention)):            
            submen.append(mention[j]['screen_name'])   
            
        mentions.append(submen)
```
